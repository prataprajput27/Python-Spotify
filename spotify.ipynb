```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set(color_codes=True)
```

```python
spotify = pd.read_csv('spotify.csv',encoding = "ISO-8859-1")
```

```python
spotify.head()
```

```python
spotify.info()
```

```python
sns.barplot(spotify['Energy'], spotify['Danceability'])
```

```python
sns.distplot(spotify['Loudness..dB..'])
```

```python
sns.distplot(spotify['Length.'], kde=False, rug=True);
```

```python
sns.jointplot(spotify['Acousticness..'], spotify['Speechiness.'])
```

```python
sns.jointplot(spotify['Acousticness..'], spotify['Speechiness.'], kind="hex")
```

```python
sns.jointplot(spotify['Acousticness..'], spotify['Speechiness.'], kind="kde")
```

```python
sns.pairplot(spotify[['Beats.Per.Minute', 'Energy', 'Danceability']])
```

```python
sns.stripplot(spotify['Track.Name'], spotify['Speechiness.'])
```

```python
sns.swarmplot(spotify['Energy'], spotify['Danceability'])
```

```python
sns.boxplot(spotify['Energy'], spotify['Danceability'], hue=spotify['Track.Name'])

import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(10,50)
```
